Q1. Convert a 1D array to a 2D array with 2 rows

```python
import numpy as np

arr = np.arange(10)
result = arr.reshape(5, 2)
print(result)
```

Q2. Get the common items between a and b

```python
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])

common = np.intersect1d(a, b)
print(common)
```

Q3. Get all items between 5 and 10 from a

```python
a = np.array([2, 6, 1, 9, 10, 3, 27])
result = a[(a >= 5) & (a <= 10)]
print(result)
```

Q9. Line plot

```python
import matplotlib.pyplot as plt

x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot")
plt.show()
```

Q10. Plot x, x², x³, x⁴ with styles

```python
x = np.arange(0., 6., 0.3)

plt.plot(x, x, 'ro', label='x')
plt.plot(x, x**2, 'bs', label='x^2')
plt.plot(x, x**3, 'g', label='x^3')
plt.plot(x, x**4, ':', label='x^4')

plt.axis([0, 6, 0, 125])
plt.legend()
plt.title("Multiple Powers of x")
plt.show()
```

Q11. Bar plot – Height Comparison

```python
height = [179, 155, 191, 152, 188, 177]
names = ['Q', 'W', 'E', 'R', 'T', 'Y']

plt.bar(names, height)
plt.title("Height Comparison")
plt.ylabel("Height (cm)")
plt.show()
```

Q12. Histogram with different bins

```python
x = np.random.randn(100000)

# 10 bins
plt.hist(x, bins=10)
plt.title("Histogram - 10 Bins")
plt.show()

# 20 bins
plt.hist(x, bins=20)
plt.title("Histogram - 20 Bins")
plt.show()

# 50 bins
plt.hist(x, bins=50)
plt.title("Histogram - 50 Bins")
plt.show()
```
